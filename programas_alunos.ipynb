```python
import pandas as pd
from datetime import datetime

# Lista de atividades pr√©-definidas
atividades_possiveis = [
    "Prova de Matem√°tica",
    "Entrega de Trabalho de Hist√≥ria",
    "Revis√£o para Prova de F√≠sica",
    "Estudo para Exame de Qu√≠mica",
    "Leitura de Textos de Literatura",
    "Elabora√ß√£o de Resumo de Biologia"
]

# Fun√ß√£o para exibir op√ß√µes e permitir escolha ou adicionar nova atividade
def escolher_atividade():
    print("\nEscolha uma atividade para incluir na agenda:")
    for i, atividade in enumerate(atividades_possiveis, 1):
        print(f"{i}. {atividade}")
    print(f"{len(atividades_possiveis) + 1}. Outra (digite sua pr√≥pria atividade)")

    while True:
        try:
            escolha = int(input("\nDigite o n√∫mero da atividade que deseja escolher: "))
            if 1 <= escolha <= len(atividades_possiveis):
                return atividades_possiveis[escolha - 1]
            elif escolha == len(atividades_possiveis) + 1:
                nova_atividade = input("Digite a nova atividade: ").strip()
                if nova_atividade:
                    return nova_atividade
                else:
                    print("Atividade inv√°lida. Tente novamente.")
            else:
                print("N√∫mero inv√°lido. Escolha um n√∫mero dentro do intervalo.")
        except ValueError:
            print("Entrada inv√°lida. Digite um n√∫mero.")

# Fun√ß√£o para criar a agenda de tarefas
def criar_agenda_tarefas():
    # Inicializando uma lista para armazenar as tarefas
    agenda = []
    hoje = datetime.now()

    print("\nCrie sua agenda de tarefas. Quando terminar, digite 'fim'.")

    while True:
        # Solicitar a data da tarefa
        data_input = input("\nDigite a data para a tarefa (formato DD/MM/AAAA) ou 'fim' para encerrar: ").strip()
        if data_input.lower() == 'fim':
            break

        # Validar a data
        try:
            data_tarefa = datetime.strptime(data_input, "%d/%m/%Y")
        except ValueError:
            print("Data inv√°lida. Tente novamente.")
            continue

        # Verificar se a data da tarefa √© no futuro
        if data_tarefa <= hoje:
            print("A data da tarefa deve ser no futuro. Tente novamente.")
            continue

        # Escolher a atividade
        atividade = escolher_atividade()

        # Calcular dias dispon√≠veis e sugerir horas di√°rias (assumindo 2 horas por dia por padr√£o)
        dias_disponiveis = (data_tarefa - hoje).days
        horas_por_dia = max(2, 10 // dias_disponiveis)  # Pelo menos 2 horas por dia, ajustado para at√© 10 horas totais

        # Adicionar a tarefa √† agenda
        agenda.append({
            "Data Tarefa": data_tarefa.strftime("%d/%m/%Y"),
            "Atividade": atividade,
            "Dias Dispon√≠veis": dias_disponiveis,
            "Horas por Dia Sugeridas": horas_por_dia
        })

    # Retornar a agenda como um DataFrame
    return pd.DataFrame(agenda)

# Criar a agenda
agenda_tarefas = criar_agenda_tarefas()

# Exibir a agenda gerada
print("\nAgenda de Tarefas:")
print(agenda_tarefas.to_string(index=False))
```
